# Новое Python API

### `MasterComponent` скрыт от пользователя

Master Component — сущность, тяжелая для понимания не-разработчика BigARTM. Кажется, возможность обучать несколько моделей по одному проходу по корпусу — не слишком востребованная особенность, ради простоты и понятности интерфейса, от нее можно отказаться. Без этой фичи можно считать, что для одной модели всегда создается один `MasterComponent`. 

Предлагается рассматривать более простой объект — контейнер модели, как estimator в SciKit-Learn.

### Тематическая модель и контейнер модели

Следует различать две сущности:
- **контейнер модели**: мутабельный объект, которому соответствует запущенный `MasterComponent` в котором живет тематическая модель, регуляризаторы и т.д., путем вызова C API можно производить итерации EM-алгоритма над моделью, содержащейся в контейнере
- **тематическая модель**: суть матрица $\Phi$ и соответствующая ей вероятностная модель смеси дискретных распределений

В BigARTM эти две сущности называются Model и TopicModel, что запутывает. Принимаются предложения по переименованию.

Некоторые другие обозначения вроде `class_id` (модальность) также вызывают вопросы, над ними тоже можно подумать.

### Контейнер модели и типичный воркфлоу

Класс `ArtmModel` по аналогии с estimator в SciKit-Learn.




Создаем контейнер модели:

```python
model = ArtmModel(num_processors=8, topics_count=1000)
```

Инициализируем модель (несколько способов):

```python
# по сгенерированной матрице
Phi0 = random_phi(topics_count, num_tokens)
model.initialize(Phi0, tokens=tokens)

# по батчам, используя некоторую схему инициализации
model.initialize_with_batches(batches, scheme='unigram')
```

Использовать [properties](https://docs.python.org/2/library/functions.html#property) в Python для того чтобы пользователь мог менять параметры модели как обычные поля объекта, напрмер:

```python
model.inner_iterations_count = 20 # присваивание изменяет сообщение ModelConfig и запускает Reconfigure
```

Делаем несколько итераций EM-алгоритма:

```python
model.fit(batches, num_iters=3, rho=0.2)
```

Добавляем регуляризацию:

```python
model.regularizers.add('smooth_words', SmoothRegularizer(class_id='words', tau=0.1))
model.regularizers.add('labels', LabelRegularizer(class_id='label', tau=10.0))
```

Еще поитерируем:

```python
model.fit(batches, num_iters=10)
```

Посмотрим что получилось:

```python
model.show_topics()
# в ноутбук выводится интерактивный просмотрщик тем
```

```python
plot(model.iterations.perplexity)
# график перплексии по итерациям
```

Что-то мало сглаживания...:

```python
model.regularizers['smooth_words'].tau = 100.0
```

```python
model.fit(batches, num_iters=5)
```

И еще посмотрим что получилось, но уже при помощи другой визуализции:

```python
model.show_topics(kind='termite')
```

Ура, получилось хорошее. Сохраним тематическую модель до лучших времен:

```python
model.save('our_topic_model')
```

Или даже экспортируем ее в CSV:

```python
model.to_csv('phi_matrix.csv')
```

Теперь загрузим сохраненную модель:

```python
model.load('our_topic_model')
```
